##### Dealing with outliers:


It doesn't make sense to have negative values for pays or benefits, so I decided to delete the rows containing negative values
But we can't decide if the maximum salaries should concidered as an outlier or not so I will keep them

```python
# Count rows with negative values
num_Negative_TotalPay=df[df['TotalPay'] < 0].shape[0]
num_Negative_BasePay=df[df['BasePay'] < 0].shape[0]
num_Negative_OvertimePay=df[df['OvertimePay'] < 0].shape[0]
num_Negative_OtherPay=df[df['OtherPay'] < 0].shape[0]
num_Negative_Benefits=df[df['Benefits'] < 0].shape[0]
num_Negative_TotalPayBenefits=df[df['TotalPayBenefits'] < 0].shape[0]

print(f"num_Negative_TotalPay: {num_Negative_TotalPay}")
print(f"num_Negative_BasePay: {num_Negative_BasePay}")
print(f"num_Negative_OvertimePay: {num_Negative_OvertimePay}")
print(f"num_Negative_OtherPay: {num_Negative_OtherPay}")
print(f"num_Negative_Benefits: {num_Negative_Benefits}")
print(f"num_Negative_TotalPayBenefits: {num_Negative_TotalPayBenefits}")
```

```python
# Drop rows with negative values
df = df[df['TotalPay'] >= 0]
df = df[df['BasePay'] >= 0]
df = df[df['OvertimePay'] >= 0]
df = df[df['Benefits'] >= 0]
df = df[df['OtherPay'] >= 0]
df = df[df['TotalPayBenefits'] >= 0]
```

### 3-Data Cleaning: Handle missing data by suitable method with explain why you use it.


---



```python
df.isnull().sum()
```

Drop columns which totally full of missing values (Notes, Status)

```python
df.drop(columns=["Notes","Status"],inplace=True)
```

```python
df.drop(columns=["Id","Agency"],inplace=True) #not necessary for analysis
```

```python
df.shape
```

Drop null values in (OvertimePay,OtherPay)

Explanation: small neglected number of missing values compared to total

```python
df.dropna(subset=["OvertimePay","OtherPay"],inplace=True)
df.shape
```

I noticed that there are some rows in which all values are zeros or missing.
so, I decided to drop these rows (26 rows)

```python
# Drop rows where all specified columns are either zeros or NaNs
df = df[~((df['BasePay'].eq(0) | df['BasePay'].isna()) &
          (df['OvertimePay'].eq(0) | df['OvertimePay'].isna()) &
          (df['OtherPay'].eq(0) | df['OtherPay'].isna()) &
          (df['Benefits'].eq(0) | df['Benefits'].isna()) &
          (df['TotalPay'].eq(0) | df['TotalPay'].isna()) &
          (df['TotalPayBenefits'].eq(0) | df['TotalPayBenefits'].isna()))]
df.shape
```

```python
df.tail(100)
```

```python
df.isnull().sum()
```

```python
# show the rows where BasePay is NaN
df[df['BasePay'].isna()]
```

Fill null values in (BasePay)

Explanation:
BasePay = TotalPay - OtherPay + OvertimePay

```python
basePay = df['TotalPay']-(df['OtherPay']+df['OvertimePay'])
df['BasePay'].fillna(basePay, inplace=True)
```

```python
benefits = df['TotalPayBenefits']-df['TotalPay']
df['Benefits'].fillna(round(benefits), inplace=True)
```

### 4- Basic Data Visualization: Create histograms or bar charts to visualize the distribution of salaries, and use pie charts to represent the proportion of employees in different departments.


---



```python
import matplotlib.pyplot as plt

plt.hist(df['TotalPay'], bins=30, color='skyblue', edgecolor='black')
plt.ylim(0, 30000)
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
department_counts = df['JobTitle'].value_counts()
maxCountsDepts = department_counts.head(12)

otherDepts=department_counts[12:].sum()

newDeptCounts=maxCountsDepts.append(pd.Series({"Other departments" :otherDepts}))

# Plot a pie chart
plt.pie(newDeptCounts, labels=newDeptCounts.index, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

###5- Grouped Analysis: Group the data by one or more columns and calculate summary statistics for each group, and compare the average salaries across different groups.


---



Here I grouped the data based on the job title to show the average salaries for each job

```python
grouped_data = df.groupby('JobTitle')

# Calculate the average salary for each Job group
average_salaries = grouped_data['TotalPay'].mean()

# Print the summary statistics
print(average_salaries.sort_values(ascending=False))
```


###6- Simple Correlation Analysis: Identify any correlation between salary and another numerical column, and plot a scatter plot to visualize the relationship.


---



```python
df.columns
```

```python
import seaborn as sns

numCols=df.select_dtypes(include=['float64', 'int64']).columns

# Create a correlation matrix
correlation_matrix = df[numCols].corr()

# Create a heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap between TotalPay and Numerical Columns')
plt.show()
```

```python
# Create a grid of subplots
num_plots = len(numCols)
num_cols = 4 # Number of columns in the grid
num_rows = (num_plots + num_cols - 1) // num_cols  # Calculate the number of rows needed

fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(15, 5 * num_rows))

# Plot each scatter plot in a separate subplot
for i, col in enumerate(numCols):
    row_index = i // num_cols
    col_index = i % num_cols
    sns.scatterplot(data=df, x='TotalPay', y=col, ax=axes[row_index, col_index])
    axes[row_index, col_index].set_title(f'Scatter Plot: TotalPay vs {col}')
    axes[row_index, col_index].set_xlabel('TotalPay')
    axes[row_index, col_index].set_ylabel(col)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```
